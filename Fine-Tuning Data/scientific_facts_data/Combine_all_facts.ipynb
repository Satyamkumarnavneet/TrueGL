### Merge all the data we have together, so that we are ready for fine-tuning

```python
import pandas as pd
```

```python
general_facts = pd.read_csv('general_facts.csv')
general_facts
```

```python
#general_facts['label']  = general_facts['label'].astype('float')
general_facts
general_facts.columns = ['statement', 'labels']
```

```python
fever_scifact_combined = pd.read_csv('fever_scifact_combined.csv')
fever_scifact_combined
```

```python
fever_scifact_combined['label'].unique()
```

```python
fever_scifact_combined.loc[fever_scifact_combined['label'] == 'SUPPORTS', 'label'] = 1
fever_scifact_combined.loc[fever_scifact_combined['label'] == 'REFUTES', 'label'] = 0
fever_scifact_combined.loc[fever_scifact_combined['label'] == 'NOT ENOUGH INFO', 'label'] = 2 # let it be like this for now
fever_scifact_combined
```

```python
#fever_scifact_combined['label'] = fever_scifact_combined['label'].astype(float)
fever_scifact_combined.columns = ['statement', 'labels']
fever_scifact_combined
```

```python
all_statements = pd.concat([general_facts, fever_scifact_combined], ignore_index=True)
all_statements
```

```python
# shuffle the dataframe
# to ensure that the training and test sets are not biased
all_statements = all_statements.sample(frac=1, random_state=42).reset_index(drop=True)
all_statements
```

```python
all_statements.to_csv('all_statements.csv', index=False)
```
