### Combine the data to make the final dataset for fine-tuning

```python
import pandas as pd
```

### For now we have three datasets:

Two of them have binary labels (either true or false) and one has more specific labels (half-true, mostly-true, false, etc.). For now it is suggested to focus on the binary classification, since the more detailed dataset has far less observations, and it might introduce the additional complexity. The model can always be further fine-tuned on it later

```python
fake_news_all = pd.read_csv('fake_news_all.csv')
fake_news_all
```

```python
welfake_data = pd.read_csv('welfake_data.csv')
welfake_data
```

```python
all_data = pd.concat([fake_news_all, welfake_data])
all_data.index = range(len(all_data))
all_data
```

```python
# ensure there are no duplicates
all_data = all_data.drop_duplicates(subset=['text'], keep='first')
all_data.index = range(len(all_data))
all_data
```

**So actually there were duplicates. Anyway, we still have a lot of training samples for fine-tuning the data.**

### Convert the data to the correct format for the fine-tuning

+ Merge the title and the text column for simplicity

+ Change the column name so that it is consistent with the format of the other dataset, and they satisfy the standards

```python
all_data['statement'] = all_data['title'] + ': ' + all_data['text']
all_data = all_data[['statement', 'label']]  # Keep only the necessary columns
all_data.rename(columns={'label': 'labels'}, inplace=True)
all_data
```
