### fake_news dataset

Available at https://huggingface.co/datasets/GonzaloA/fake_news

Also can be checked at https://www.kaggle.com/datasets/aadyasingh55/fake-news-classification (and there is also some code, hence, might be helpful for the analysis)

```python
#! git clone https://hf-mirror.com/datasets/GonzaloA/fake_news
```

```python
import pandas as pd
```

```python
# 1 - true, 0 - false
fake_news_train = pd.read_csv('fake_news/train.csv', on_bad_lines='skip', sep=';')
fake_news_train = fake_news_train.drop('Unnamed: 0', axis=1)
fake_news_train
```

```python
fake_news_eval = pd.read_csv('fake_news/evaluation.csv', on_bad_lines='skip', sep=';')
fake_news_eval = fake_news_eval.drop('Unnamed: 0', axis=1)
fake_news_eval
```

```python
fake_news_test = pd.read_csv('fake_news/test.csv', on_bad_lines='skip', sep=';')
fake_news_test = fake_news_test.drop('Unnamed: 0', axis=1)
fake_news_test
```

```python
fake_news_all = pd.concat([fake_news_train, fake_news_eval, fake_news_test])
fake_news_all.index = range(len(fake_news_all))
fake_news_all
```

```python
# pretty balanced
fake_news_all.label.value_counts()
```

```python
fake_news_all.to_csv('fake_news_all.csv', index=False)
```

### LIAR Dataset

Obtained from https://www.kaggle.com/datasets/doanquanvietnamca/liar-dataset/data

Note that the listed usability is not high, but probably it is still worth checking. And also there is some code related to this dataset, maybe can find something for fine-tuning

```python
liar_columns = [
    "ID", "label", "statement", "subjects", "speaker", "speaker_job_title",
    "state_info", "party_affiliation", "barely_true_counts", "false_counts",
    "half_true_counts", "mostly_true_counts", "pants_on_fire_counts", "context"
]

liar_train = pd.read_csv('LIAR_DATASET/train.tsv', sep='\t', header=None, names=liar_columns)
liar_train
```

```python
liar_valid = pd.read_csv('LIAR_DATASET/valid.tsv', sep='\t', header=None, names=liar_columns)
liar_valid
```

```python
liar_test = pd.read_csv('LIAR_DATASET/test.tsv', sep='\t', header=None, names=liar_columns)
liar_test
```

```python
liar_all = pd.concat([liar_train, liar_valid, liar_test])
liar_all.index = range(len(liar_all))
liar_all
```

```python
# this dataset is good because the truth is divided into categories
liar_all.label.value_counts()
```

##### I think we can keep only the statement and label (at least for now)

```python
liar_all = liar_all[['statement', 'label']]
liar_all
```

```python
liar_all.to_csv('liar_all.csv', index=False)
```

### WELFake

Obtained from https://www.kaggle.com/datasets/saurabhshahane/fake-news-classification

Usability is listed as 10 / 10

Also note that there is a lot of code with analysis here. Hence, maybe it can be useful, and maybe we can find the code for fine-tuning the model using this data there

```python
# pretty balanced
# still only two labels
welfake_data.label.value_counts()
```
